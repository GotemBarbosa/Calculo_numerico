# Metodo dos minimos quadradados

```python
def mmq(x, y, k):
    n = len(x) # quantidade de pontos
    X = np.vander(x) #Matriz de Vandermonde
    X = X[:, n-k-1:n]  # Seleciona apenas as Ãºltimas k colunas
    a = np.linalg.solve(X.T @ X, X.T @ y)  # Resolve o sistema normal
    return a
    
```

```python
def mmq(x, y, k):
    n = len(x) # quantidade de pontos
    X = np.vander(x) #Matriz de Vandermonde
    X = X[:, n-k-1:n]  # Seleciona apenas as Ãºltimas k colunas
    Q,R = np.linalg.qr(X)  # DecomposiÃ§Ã£o QR
    a = np.linalg.solve(R, Q.T @ y)  # Resolve o sistema normal
    return a



```
