# Interpolacao com metodo de lagrange

```python
def lagrange(x_i, y_i, x): #os pontos de xi, seus yi correspondentes
    
    L = np.ones((np.size(x_i), np.size(x))) #inicializa Lk

    for i in range (np.size(x_i)): #para os Lk
        for k in range (np.size(x_i)):
            if (i != k):
                L[i,:] = (L[i,:]*(x-x_i[k])/(x_i[i]-x_i[k])) # faz Lk usando os xi por meio da expressao
    y = y_i.dot(L) #soma dos Lk*yk
    return y


x = np.linspace(-2,5, num=50, endpoint=True) #cria os pontos xi
x_i = np.array([-2, 0, 3, 5]) #pontos xi
y_i = np.array([3,-2, 4, 2]) #pontos yi
y = lagrange(x_i, y_i, x) #chama a funcao

import matplotlib.pyplot as plt
plt.plot(x, y, label='Lagrange')
plt.scatter(x_i, y_i, color='red', label='Pontos dados')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interpolação de Lagrange')
plt.legend()
plt.grid()
plt.show()

```
